```python
# Task-2: Composite Score Ranking
# Q1: Using the SouthAsia DataFrame, create a new column called Composite Score that combines the following metrics:
# Composite Score = 0.40GDP per Capita + 0.30Social Support + 0.30Healthy Life Expectancy
# Explanation: We are creating a new metric to rank the countries based on these important factors.
# This composite score will give a weighted average of the country's GDP, Social Support, and Life Expectancy.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the South Asia dataset (already filtered and saved previously)
south_asia_df = pd.read_csv('/content/drive/MyDrive/south_asian_countries.csv')

# Q2: Rank the South Asian countries based on the Composite Score in descending order.
# Explanation: We will rank the countries based on the calculated Composite Score, with the highest-ranked countries at the top.

# Create the Composite Score
south_asia_df['Composite Score'] = (0.40 * south_asia_df['Log GDP per capita'] +
                                     0.30 * south_asia_df['Social support'] +
                                     0.30 * south_asia_df['Healthy life expectancy'])

# Sort the countries by Composite Score in descending order
south_asia_df = south_asia_df.sort_values(by='Composite Score', ascending=False)

# Q3: Visualize the top 5 countries using a horizontal bar chart showing the Composite Score.
# Explanation: We will use a bar chart to display the top 5 countries based on the Composite Score.
plt.figure(figsize=(10,6))
sns.barplot(x='Composite Score', y='Country name', data=south_asia_df.head(5), palette='viridis')
plt.title('Top 5 South Asian Countries by Composite Score')
plt.xlabel('Composite Score')
plt.ylabel('Country name')
plt.show()

# Q4: Discuss whether the rankings based on the Composite Score align with the original Score- support your discussion with some visualization plot.
# Explanation: In this step, we visualize the relationship between the Composite Score and the original Happiness Score (Score) to check if both rankings align.

# Visualize the relationship between Composite Score and Happiness Score using a scatter plot
plt.figure(figsize=(10,6))
sns.scatterplot(x='Composite Score', y='score', data=south_asia_df, color='blue')
plt.title('Composite Score vs Happiness Score')
plt.xlabel('Composite Score')
plt.ylabel('Happiness Score')
plt.show()
```

```python
# Q3.1: Identify outlier countries in South Asia based on their Score and GDP per Capita.
# Explanation: Outliers are extreme data points that differ significantly from the majority of the data.
# We will use the 1.5  IQR rule to detect these outliers.

# 1. Calculate IQR for Score and GDP per Capita
score_q1 = south_asia_df['score'].quantile(0.25)
score_q3 = south_asia_df['score'].quantile(0.75)
score_iqr = score_q3 - score_q1

gdp_q1 = south_asia_df['Log GDP per capita'].quantile(0.25)
gdp_q3 = south_asia_df['Log GDP per capita'].quantile(0.75)
gdp_iqr = gdp_q3 - gdp_q1

# 2. Calculate lower and upper bounds for outliers using the 1.5  IQR rule
score_lower_bound = score_q1 - 1.5 * score_iqr
score_upper_bound = score_q3 + 1.5 * score_iqr

gdp_lower_bound = gdp_q1 - 1.5 * gdp_iqr
gdp_upper_bound = gdp_q3 + 1.5 * gdp_iqr

# 3. Identify outliers based on the bounds
outliers = south_asia_df[(south_asia_df['score'] < score_lower_bound) |
                          (south_asia_df['score'] > score_upper_bound) |
                          (south_asia_df['Log GDP per capita'] < gdp_lower_bound) |
                          (south_asia_df['Log GDP per capita'] > gdp_upper_bound)]

# Q3.2: Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers in a different color.
# Explanation: We will create a scatter plot showing GDP per Capita vs. Happiness Score, with outliers highlighted in red.

plt.figure(figsize=(10,6))
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_df, label='Normal Data', color='blue')

# Highlighting outliers in red
sns.scatterplot(x='Log GDP per capita', y='score', data=outliers, color='red', label='Outliers')

plt.title('Outliers in South Asia based on GDP per Capita and Score')
plt.xlabel('Log GDP per capita')
plt.ylabel('Happiness Score')
plt.legend()
plt.show()

# Q3.3: Discuss the characteristics of these outliers and their potential impact on regional averages.
# Explanation: Outliers can significantly affect the overall averages and metrics used for the region. In this case,
# outliers based on GDP per Capita and Happiness Score may skew the results and provide misleading insights for the region.
# We will now display the outliers for a closer examination.

outliers
```

```python
# Q4.1: Calculate the Pearson correlation between 'Freedom to make life choices' and 'Generosity' with the Score.
# Explanation: Pearson correlation measures the linear relationship between two variables.

# Calculate Pearson correlation between 'Freedom to make life choices' and 'Score'
freedom_corr = south_asia_df['Freedom to make life choices'].corr(south_asia_df['score'])

# Calculate Pearson correlation between 'Generosity' and 'Score'
generosity_corr = south_asia_df['Generosity'].corr(south_asia_df['score'])

# Display the correlation values
print(f"Pearson correlation between 'Freedom to make life choices' and 'Score': {freedom_corr}")
print(f"Pearson correlation between 'Generosity' and 'Score': {generosity_corr}")

# Q4.2: Create scatter plots with trendlines for these metrics against the Score.
# Explanation: Scatter plots with trendlines to visualize the relationship between these metrics and the Score.

# Plot for 'Freedom to make life choices' vs 'Score'
plt.figure(figsize=(10, 6))
sns.regplot(x='Freedom to make life choices', y='score', data=south_asia_df, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.title('Scatter Plot: Freedom to make life choices vs Happiness Score')
plt.xlabel('Freedom to make life choices')
plt.ylabel('Happiness Score')
plt.show()

# Plot for 'Generosity' vs 'Score'
plt.figure(figsize=(10, 6))
sns.regplot(x='Generosity', y='score', data=south_asia_df, scatter_kws={'color':'green'}, line_kws={'color':'red'})
plt.title('Scatter Plot: Generosity vs Happiness Score')
plt.xlabel('Generosity')
plt.ylabel('Happiness Score')
plt.show()

# Q4.3: Identify and discuss the strongest and weakest relationships between these metrics and the Score for South Asian countries.
# Explanation: After viewing the correlation values and scatter plots, we will discuss the strongest and weakest relationships.

# Discussion based on correlation values:
if abs(freedom_corr) > abs(generosity_corr):
    print("The strongest relationship is between 'Freedom to make life choices' and 'Score'.")
else:
    print("The strongest relationship is between 'Generosity' and 'Score'.")

if freedom_corr == 0:
    print("There is no correlation between 'Freedom to make life choices' and 'Score'.")
if generosity_corr == 0:
    print("There is no correlation between 'Generosity' and 'Score'.")
```

```python
# Q1: Add a new column, "GDP-Score Gap", which is the difference between GDP per Capita and the Score for each South Asian country.
# Explanation: This will allow us to see how much the GDP per capita deviates from the happiness score for each country.

south_asia_df['GDP-Score Gap'] = south_asia_df['Log GDP per capita'] - south_asia_df['score']


# Q2: Rank the South Asian countries by this gap in both ascending and descending order.
# Explanation: We will rank the countries based on the gap to identify which countries have the largest positive and negative gaps
# between GDP per capita and their happiness score.

# Ascending order (smallest gap to largest)
ascending_gap = south_asia_df.sort_values(by='GDP-Score Gap', ascending=True)

# Descending order (largest gap to smallest)
descending_gap = south_asia_df.sort_values(by='GDP-Score Gap', ascending=False)

# Display the first 10 countries with the largest positive gap (descending)
print("Top 10 Countries with Largest Positive GDP-Score Gap:")
print(descending_gap[['Country name', 'GDP-Score Gap']].head(10))

# Display the first 10 countries with the largest negative gap (ascending)
print("\nTop 10 Countries with Largest Negative GDP-Score Gap:")
print(ascending_gap[['Country name', 'GDP-Score Gap']].head(10))


# Q3: Highlight the top 3 countries with the largest positive and negative gaps using a bar chart.
# Explanation: Visualizing the top 3 countries with the largest positive and negative gaps using a bar chart can give us insights into the
# disparities between GDP per capita and happiness.

# Top 3 countries with the largest positive gaps
top_3_positive = descending_gap.head(3)

# Top 3 countries with the largest negative gaps
top_3_negative = ascending_gap.head(3)

# Combine the top positive and negative gaps
top_gaps = pd.concat([top_3_positive, top_3_negative])

# Plotting the bar chart
plt.figure(figsize=(10, 6))

# Creating the bar chart
sns.barplot(x='GDP-Score Gap', y='Country name', data=top_gaps, palette='coolwarm')

# Adding title and labels
plt.title('Top 3 Positive and Negative GDP-Score Gaps')
plt.xlabel('GDP-Score Gap')
plt.ylabel('Country name')

# Display the chart
plt.show()


# Q4: Analyze the reasons behind these gaps and their implications for South Asian countries.
# Explanation: The analysis will discuss the characteristics of the countries with the largest positive and negative gaps.
# For instance:
# - Positive Gaps: Countries with high GDP but low happiness scores may be facing high income inequality, political instability,
# or other issues that hinder happiness, such as lack of social services or a poor healthcare system.
# - Negative Gaps: Countries with low GDP but high happiness scores could have strong community support, effective social safety nets,
# or a high sense of national well-being that contributes to a higher score.

# Analysis discussion:
# The positive gaps typically show that economic growth does not always translate into high happiness. This can be due to factors like
# inequality, corruption, or lack of social and healthcare systems. Countries like Sri Lanka, despite being lower in GDP per capita,
# might score higher in happiness due to strong community ties, lower inequality, or political stability.
# Conversely, countries with high GDP and low happiness scores, such as India and Pakistan, may struggle with internal inequality,
# corruption, and lack of public health infrastructure, which contribute to their relatively low happiness despite higher economic output.

```
